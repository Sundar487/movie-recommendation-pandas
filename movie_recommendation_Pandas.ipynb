```python
from collections import Counter
```

```python
top_movies = Counter(df["movie_id"]).most_common(10)
```

```python
def get_recommendation(user1):
  watched_movies = df[df["user_id"]==user1]["movie_id"]
  return [x for x in movies_suggestion if x not in watched_movies]
```
